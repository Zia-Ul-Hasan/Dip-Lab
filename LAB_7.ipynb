GREY SLICING



````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def f(x):
    # Use numpy's vectorized operations for each condition
    y = np.where(x <= 0.25, 1 - x, np.where(x <= 0.5, 1, x))
    return y

# Plotting the function for a range of values to visualize the transformation
x = np.linspace(0, 1, 500)
y = f(x)
plt.plot(x, y)
plt.xlabel("Input Intensity (x)")
plt.ylabel("Output Intensity (f(x))")
plt.title("Graph of f(x) for Image Intensity Transformation")
plt.show()
````

````python
from google.colab.patches import cv2_imshow
image = cv2.imread('/content/bg_1.png',cv2.IMREAD_GRAYSCALE)

normalized = image/255.0

transformed = f(normalized)
tn = (transformed *255.0).astype(np.uint8)
cv2_imshow(image)
cv2_imshow( tn)
````

HISTOGRAM EQUALIZATION

````python
import matplotlib.pyplot as plt
image2= cv2.imread('/content/hist_eq.png', cv2.IMREAD_GRAYSCALE)

histogram,bins=np.histogram(image2.flatten(),256,[0, 256])

norm_hist = histogram/histogram.sum()

cdf = norm_hist.cumsum()
norm_cdf = (cdf*256).astype('uint8')

eq_image = norm_cdf[image2]
hist_eq,bins=np.histogram(eq_image.flatten(),256,[0, 256])

cv2_imshow(image2)
plt.plot(histogram)
cv2_imshow(eq_image)
plt.plot(hist_eq)
````

CONTRAST STRECHING


````python
# Corrected code
image3 = cv2.imread('/content/maxresdefault.png', cv2.IMREAD_GRAYSCALE)

min_v = np.min(image3)
max_v = np.max(image3)
streched_formula = (image3 - min_v) * (255 / (max_v - min_v))
streched_formula = streched_formula.astype(np.uint8)

# Display the contrast-stretched image
cv2_imshow(streched_formula)
````
