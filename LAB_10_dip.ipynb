# **TASK 1**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/content/image_2024-11-28_232653828.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.show()

kernel = np.array([[-1, -1, -1], [2, 2, 2], [-1, -1, -1]])
filtered_image = cv2.filter2D(image, -1, kernel)
plt.imshow(filtered_image, cmap='gray')
plt.title("Horizontal Edges")
plt.show()
```

# TASK 2

```python
image2 = cv2.imread('/content/image_2024-11-28_232831867.png', cv2.IMREAD_GRAYSCALE)
sobel_x = cv2.Sobel(image2, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image2, cv2.CV_64F, 0, 1, ksize=3)

plt.subplot(1, 2, 1)
plt.imshow(sobel_x, cmap='gray')
plt.title("Horizontal Edges")

plt.subplot(1, 2, 2)
plt.imshow(sobel_y, cmap='gray')
plt.title("Vertical Edges")

plt.show()
```

```python
combined_edges = cv2.magnitude(sobel_x, sobel_y)
plt.imshow(combined_edges, cmap='gray')
plt.title("Combined Edges")
plt.show()
```

# **TASK 3**

```python
image3 = cv2.imread('/content/image_2024-11-28_233302415.png', cv2.IMREAD_GRAYSCALE)
blurred = cv2.GaussianBlur(image3, (5, 5), 0)
edges = cv2.Canny(blurred, 100, 200)
plt.imshow(edges, cmap='gray')
plt.title("Detected Edges")
plt.show()
```
